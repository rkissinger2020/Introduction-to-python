```python
#mini dictionary

student = {
    "name": "Alice",
    "age": 14,
    "grade": "B"
}


print("Student name:", student["name"])


student["grade"] = "A"


student["school"] = "Greenwood High"


del student["age"]

print("Updated student dictionary:", student)


person = {"name": "Bob", "age": 30, "city": "Nairobi"}

print("value pairs:")
for key, value in person.items():
    print(f"{key}: {value}")


print("Keys:")
for key in person:
    print(key)


print("Values:")
for value in person.values():
    print(value)


email = person.get("email", "Not provided")
print("Email:", email)


if "city" in person:
    print("Key 'city' exists in the dictionary.")


person.update({"city": "Mombasa", "email": "bob@example.com"})
print(" Updated person dictionary:", person)


students = [
    {"name": "Liam", "age": 15, "grade": 85},
    {"name": "Emma", "age": 14, "grade": 92},
    {"name": "Noah", "age": 16, "grade": 78}
]


print("Student names and grades:")
for student in students:
    print(f"{student['name']} - Grade: {student['grade']}")


total = 0
for student in students:
    total += student["grade"]
average = total / len(students)
print("Average grade: {average}")


words = ["apple", "banana", "apple", "cherry", "banana", "apple"]
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1

print("Word count:", word_count)


print("Phonebook program:")
phonebook = {}

while True:
    action = input("Enter 'add', 'lookup', or 'quit': ").strip().lower()

    if action == "add":
        name = input("Enter name: ")
        number = input("Enter phone number: ")
        phonebook[name] = number
    elif action == "lookup":
        name = input("Enter name to lookup: ")
        print(phonebook.get(name, "Not found"))
    elif action == "quit":
        print("Exiting phonebook.")
        break
    else:
        print("Invalid option.")
```
